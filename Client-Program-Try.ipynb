```python
import socket
import threading
import cv2
import numpy as np
import sys
import pickle
```

```python
#Creating Socket Program
protocol = socket.SOCK_STREAM
net_family = socket.AF_INET
s = socket.socket(net_family, protocol)
```

```python
#Connecting to the Server
serverip = "192.168.29.222"
port = 8889
s.connect((serverip, port))
print(s.recv(9))
```

```python
#Selecting the Camera Input
cap = cv2.VideoCapture(1)
```

```python
cap.isOpened()
```

```python
#Function to Send the image stream to server

def Send(s):
    while True:
        status, photo = cap.read()
        if type(photo) is type(None):
            pass
        else:
            dumped_photo = pickle.dumps(photo)
            s.sendall(dumped_photo)
    cap.release()
    
"""

def Send(s):
    while True:
        status, photo = cap.read()
        if type(photo) is type(None):
            pass
        else:
            photo_to_bytes = cv2.imencode('.jpg',photo)[1].tobytes()
            s.sendall(photo_to_bytes)
    cap.release()

"""
```
